```python
import pandas as pd
import json
from os.path import join
from os import listdir
import altair as alt
import pickle
```

```python
folder_names = ['fintech', 'investment', 'other']

rows_fin = []
rows_inv = []
for folder in folder_names:
    path = join('data', folder)
    posts_names = listdir(path)
    for post in posts_names:
        with open(join(path, post), encoding='utf-8') as f:
            if folder == 'fintech':
                r = json.load(f)
                r['fintech'] = 1
                rows_fin.append(r)
            elif folder == 'investment':
                v = json.load(f)
                v['investment'] = 1
                rows_inv.append(v)
            else:
                q = json.load(f)
                r = dict(q)
                v = dict(q)
                r['fintech'] = 0
                rows_fin.append(r)
                v['investment'] = 0
                rows_inv.append(v)
            
            
df_fin = pd.DataFrame.from_dict(rows_fin, orient='columns')
df_inv = pd.DataFrame.from_dict(rows_inv, orient='columns')
```
